```python
data = pd.read_csv("diabetes_data_upload.csv")
data.head()
```

```python
data.isna().sum()
```

At first glance it seems like there are no missing values. But the webpage states otherwise, so let's look for them. Every column has only binary values (Yes/No, Male/Female, Positive/Negative) except age. So, let's encode them to 1 and 0. 
If any problem arises during label encoding, that's because of NaNs.

```python
encoding_dict = {"Yes": 1, "No": 0, "Male": 1, "Female": 0, "Positive": 1, "Negative": 0}
for col in data.columns:
    count = 0
    try:
        data[col] = [encoding_dict[x] for x in data[col]]
    except KeyError:
        count +=1
        print(col," has a non-binary value.")
    
```

```python
data["Age"].value_counts()
```

The age column doesn't seem to have NaNs either. So, we can conclude, there are no NaN values in the dataset.

## 4. Evaluation
 - generate prediction probabilities with a decision tree and with a random forest model
 - use 5 fold cross validation for both time (so you should get 520 predictions)
 - use default parameters for both models
 - compare the two models with ROC curves
   - why does the decision tree's ROC curve looks different?
 

```python
from sklearn.model_selection import KFold
from sklearn.metrics import plot_roc_curve

kf = KFold(n_splits = 5)
plt.figure(figsize=[16,6])
i = 0
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    clf2.fit(X_train, Y_train)
    clf3.fit(X_train, Y_train)
     
    
    plt.subplot(1,2, 1)
    plot_roc_curve(clf2, X_test, Y_test, ax = plt.gca())
    
    plt.subplot(1,2, 2)
    plot_roc_curve(clf3, X_test, Y_test, ax = plt.gca())
    i +=0

plt.subplot(121)
plt.title("Decision trees",size=18)
plt.plot(np.linspace(0,1,100),np.linspace(0,1,100),color="black")
plt.grid()

plt.subplot(122)
plt.title("Random forest",size=18)
plt.plot(np.linspace(0,1,100),np.linspace(0,1,100),color="black")
plt.grid()
```

We can see, that almost every random forest is the same, while the decision trees are different. This might be because the random forest uses by default n_trees = 100, which might be too much, thus resulting in overfitting.

## 5. Tuning model
  - using 80/20% train/test split generate predictions for a random forest model
  - plot the AUC vs number of trees in the forest for both the traing and the test data
  - do we experience overfitting if we use too many trees?

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc, roc_curve
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
auc_list = []
N_TREES_list = np.arange(1,50)
for N_TREES in N_TREES_list:
    clf_temp = RandomForestClassifier(n_estimators = N_TREES)
    clf_temp.fit(X_train, Y_train)
    pred = clf_temp.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(Y_test, pred[:,1])
    auc_list.append(auc(fpr, tpr))
```

```python
plt.figure(figsize=[8,6])
plt.plot(N_TREES_list, auc_list, ls = "--", marker = "o", color="tab:purple")
plt.grid()
plt.xlabel("number of trees",size=14)
plt.ylabel("AUC score", size=14)
```

We can see, that from around 10 trees the AUC is very close to 1 with some fluctuations. From the perspective of using too many trees, [Breiman (who introduced them) claims, that random forests can't overfit by nature](https://www.stat.berkeley.edu/~breiman/RandomForests/cc_home.htm#remarks). <br>
However, any estimator can overfit, like in this case, as the variance of generalization error is approching zero while adding more trees, but the bias will not. I suspect, that the randomness in building random forests vanishes, because we're starting to cover the whole sample space.
