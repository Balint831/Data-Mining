# Solution
## 1.) Linear SVC in case of linear separation

- load the Iris dataset (can be found in sklearn API)
- scale the data and plot the petal length vs petal width in a scatterplot colored with the target
- train an SVC model with linear kernel with default parameter settings, but once with C=1 and then C=1000
- visualize the model's decision boundary and the margins based on the coefficients learnt by the model
- interpret the results, what is the role of the C hyperparameter?

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
sScaler = StandardScaler()
iris.data = sScaler.fit_transform(iris.data)
```

```python
plt.figure(figsize=[8,6])
sns.scatterplot(x = iris.data[:,2], y = iris.data[:,3], hue = iris.target_names[iris.target],palette = "spring")
plt.xlabel("petal length",size=14)
plt.ylabel("petal width",size=14)
plt.grid()
plt.title("Petal length vs petal length",size=18)
plt.legend(fontsize=14)
```

On the figure above we can see that the separate species are clumped together forming clusters, thus we could draw pretty concise decision boundaries for each species.

```python
def plot_decision_boundary( data, target, C, isIris = True):
    #fitting the classifier, calculating the margin
    svc = SVC(kernel='linear', C=C)
    
    if isIris:
        svc.fit(data[:,2:], target)
        
    else: 
        svc.fit(data,target)

    #plotting the datapoints
    if isIris:
        sns.scatterplot(x = data[:,2], y = data[:,3], hue = iris.target_names[target],palette = "spring")
        
    else:
        sns.scatterplot(x = data[:,0], y = data[:,1], hue = y,palette = "spring")
        
    #setting the variables of the plot
    w = svc.coef_[0]
    a = -w[0] / w[1]
    
    ax = plt.gca()
    xlim = ax.get_xlim()
    xx = np.linspace(xlim[0], xlim[1],1000)
    yy = a * xx - (svc.intercept_[0]) / w[1]

    # plot the parallels to the separating hyperplane that pass through the
    # support vectors (margin away from hyperplane in direction
    # perpendicular to hyperplane). This is sqrt(1+a^2) away vertically in
    # 2-d.
    margin = 1 / np.sqrt(np.sum(svc.coef_ ** 2))
    yy_down = yy - np.sqrt(1 + a ** 2) * margin
    yy_up = yy + np.sqrt(1 + a ** 2) * margin
    
    plt.plot(xx, yy, color="red")
    plt.plot(xx, yy_down, color="purple",ls = "--")
    plt.plot(xx, yy_up, color="purple", ls = "--")
    plt.fill_between(xx, yy_down, yy_up, edgecolor='none',
                     color="gray", alpha=0.4)

    if isIris:
        plt.xlabel("petal length",size=14)
        plt.ylabel("petal width",size=14)
    plt.grid()
    plt.title("Decision boundary with C = {}".format(C),size=18)
    plt.legend(fontsize=14)
    print(np.sqrt(np.sum(svc.coef_ ** 2)))
```

```python
plt.figure(figsize=[24,12])
plt.subplot(231)
plot_decision_boundary(iris.data[iris.target != 0], iris.target[iris.target != 0],1)
plt.subplot(234)
plot_decision_boundary(iris.data[iris.target != 0], iris.target[iris.target != 0],1000)

plt.subplot(232)
plot_decision_boundary(iris.data[iris.target != 1], iris.target[iris.target != 1],1)
plt.subplot(235)
plot_decision_boundary(iris.data[iris.target != 1], iris.target[iris.target != 1],1000)

plt.subplot(233)
plot_decision_boundary(iris.data[iris.target != 2], iris.target[iris.target != 2],1)
plt.subplot(236)
plot_decision_boundary(iris.data[iris.target != 2], iris.target[iris.target != 2],1000)
```

On the graph above we could see, that in the versicolor-virginica comparison the $C$ parameter played a significant role, meaning it greatly reduced the margin. With $C=1000$ the position of the threshold and the orientation changed a small amount. The second pair of graphs gave us no additional information with regards to the $C$ parameter. The <br>
Increasing $C$ regularization parameter will shrink the soft margin of the support vector classifier in most cases. Thus, the classification would give us less wrong answers, with greater certainty. Other than that we could notice a little increase in the steepness of the decision boundary. So, decreasing $C$ corresponds to more regularization.

## 2.) Linear SVC but non-linear separation

- create a dataset with the following: X, y = sklearn.datasets.make_moons(noise=0.1, random_state=0)
- perform the same steps just as in the previous exercise and use the linear kernel for  the SVC
- since linear SVC cannot do non-linear separation, you will need to do some workaround, for example adding polynomial features (3rd order would be a good choice)
- write down with your own words in few sentences how the support vector machine works 

```python
from sklearn.datasets import make_moons
X,y = make_moons(noise = 0.1, random_state = 0)
```

```python
plt.figure(figsize=[16,6])
plt.subplot(121)
plot_decision_boundary(X, y,1,isIris=False)
plt.subplot(122)
plot_decision_boundary(X, y,1000,isIris=False)
```

It is clearn from the plot above, that in the case of the crescents the linear kernel is not helpful. However, setting $C$ higher will decrease the margin, but the decision boundary is still wrong.

```python
plt.figure(figsize=[16,6])
X_poly = np.array([[x[0]**3, x[1]**3] for x in X])
plt.suptitle("With third power features",size=18)
plt.subplot(121)
plot_decision_boundary(X_poly, y,1,isIris=False)
plt.subplot(122)
plot_decision_boundary(X_poly, y,1000,isIris=False)
```

I tried transforming the input features in a way to give $\big[x[0],x[1],x^3[0],x^3[1]\big]$ to the SVM. It resulted in the graph above. On the $C=1$ graph the decision boundary somewhat separates the data, but the margin is so big, that there might be a lot of misclassifications. However with $C=1000$ the separation is relatively better, and the margin is much thinner.

## 3.) Load the dataset from 2 weeks ago and build/evaluate the SVC with default settings

Reminder:
- you need to build a classifier that predicts the probability of a sample coming from a cancerous (tumor type is normal or not) person based on the measured protein levels
- train the SVM classifier (SVC in sklearn API) on every second sample (not first 50% of the data (!), use every second line)
- generate prediction for the samples that were not used during the training

To-do now:
- build default SVC, but set it to predict probabilities
- plot the ROC curve and calculate the confusion matrix for the predictions
- do the same for the CancerSEEK predictions and compare your model's performance to CancerSEEK performance (as a reference, plot it on the same figure)
- how good is the performance of the new model?

```python
data = pd.read_excel("aar3247_Cohen_SM_Tables-S1-S11.xlsx",sheet_name="Table S6",header=2, index_col="Sample ID #")

#Drop last 4 rows, they're all just nans
data = data[:-4] 

#Dataframe mask for part 4
mask4 = (data["Tumor type"] == "Normal") | (data["Tumor type"] == "Liver")

# Convert categorical data to numerical values
y_CSEEK = data["CancerSEEK Test Result"].map({"Positive": 1, "Negative":0})
y = data["Tumor type"].map( lambda x: 0 if x == "Normal" else 1)

# Drop unnecesary columns
data = data.drop(columns = ["Patient ID #","AJCC Stage","CancerSEEK Logistic Regression Score","CancerSEEK Test Result"])
```

```python
for col in data.columns[1:]:
    data[ col ] = data[ col ].map(lambda x: float(x.lstrip("*")) if isinstance(x, str) else x)
    # When the cell value is of type string, it is because the cell has * or ** in it. 
    # If True remove * with lstrip from the left side of the cells,
    # else return original.
```

```python
num_of_na = data.isna().sum(axis=0) #Dict with number of missing features by column

[print(x,":",num_of_na[x]) 
 if num_of_na[x] != 0 
 else None 
 for x in data.columns] # Broken into multiple lines for better readability

print("\nratio of missing features in the whole table: ", data.isna().sum().sum()/data.count().sum() )
```

```python
data = data.fillna(data.mean()) # Impute missing values with the mean of the columns
data.isna().sum().sum(), y.isna().sum().sum()
```

<b>No missing values left.

```python
x_train = data[data.columns[1:]][::2] #Pick every second record.
x_test = data[data.columns[1:]][1::2]
y_train = y[::2]
y_test = y[1::2]
```

```python
from sklearn.metrics import plot_roc_curve, plot_confusion_matrix, confusion_matrix, roc_curve, auc

plt.figure(figsize=[16,12])
plt.subplot(221)
svm3 = SVC(probability = True)
svm3.fit(x_train,y_train)
y_pred = svm3.predict(x_test)
plot_roc_curve(svm3, x_test, y_test, ax = plt.gca())
plt.plot(np.linspace(0,1,100),np.linspace(0,1,100))
plt.grid()

plt.subplot(222)
plot_confusion_matrix(svm3, x_test, y_test, ax = plt.gca())
plt.suptitle("Metrics of classification - not scaled data", fontsize=22)

plt.subplot(223)
plt.title("CancerSEEK comparison",size=18)
fpr, tpr, thresholds = roc_curve(y_CSEEK[1::2], svm3.predict_proba(x_test)[:,1])
roc_auc = auc(fpr,tpr)
plt.plot(fpr,tpr, label="SVC - AUC: "+str(round(roc_auc,3)))
plt.plot(np.linspace(0,1,100),np.linspace(0,1,100))
plt.legend(loc = 4)
plt.grid()
plt.ylabel("True positive rate - wrt CSEEK",size=16)
plt.xlabel("False positive rate - wrt CSEEK",size=16)



conf_matr = pd.DataFrame(confusion_matrix(y_CSEEK[1::2],y_pred), columns = [0,1],index = [0,1])
plt.subplot(224)
plt.imshow(conf_matr)
plt.title("CancerSEEK comparison",size=18)
for i in range(2):
    for j in range(2):
        plt.text(i,j, conf_matr[i][j],ha="center", va="center", color="red", size = 18)
        
plt.xticks(np.arange(2), labels = [0,1],size=16)
plt.yticks(np.arange(2), labels = [0,1],size=16)
plt.xlabel("Predicted label",size=17)
plt.ylabel("CancerSEEK Label",size=17)
```

If we look at our model, we can see, that the most of the patients were correctly classified. It is also fair to say, that we misclassified a substantial amount of samples to be healthy that were in fact cancerous. <br>
Comparing our model with the CancerSEEK results we can say, that our model predicted much more samples to be cancerous, than CSEEK. Thus the false positive rate with respect to CancerSEEK was greater in general. There are just a small number of people who couldn't be indentified to have cancer.
